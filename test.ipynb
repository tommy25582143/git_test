```python
import networkx as nx
import matplotlib.pyplot as plt
from torch_geometric.data import HeteroData

# Create a sample HeteroData instance (re-using the structure from your example)
data = HeteroData()
data['paper'].x = [1, 2, 3, 4]  # Simplified node features for visualization
data['author'].x = [1, 2, 3]
data['paper'].title = ["Paper on GNNs", "Paper on AI", "Survey on ML", "Research on Graphs"]
data['author'].name = ["Alice", "Bob", "Charlie"]
data['author', 'writes', 'paper'].edge_index = [[0, 0, 1], [1, 2, 3]]
data['paper', 'cites', 'paper'].edge_index = [[0, 1, 1], [1, 2, 3]]
```

```python
# Assume G is already created and populated with nodes and edges
G = nx.DiGraph()

# Adding nodes with color attributes directly in NetworkX
for node_id, name in enumerate(["Alice", "Bob", "Charlie"]):  # Example names for authors
    G.add_node(f"Author_{name}", label=name, color='skyblue')
for node_id, title in enumerate(["Paper on GNNs", "Paper on AI", "Survey on ML", "Research on Graphs"]):
    G.add_node(f"Paper_{title}", label=title, color='lightgreen')

# Example edges, assuming you've added these correctly
G.add_edge("Author_Alice", "Paper_Paper on GNNs")
G.add_edge("Paper_Paper on GNNs", "Paper_Paper on AI")

# Drawing the graph
pos = nx.spring_layout(G)  # Positions for all nodes

# Draw nodes with colors
node_colors = [G.nodes[n]['color'] for n in G.nodes()]
nx.draw_networkx_nodes(G, pos, node_color=node_colors)

# Draw edges and labels
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos)

# Show the plot
plt.title('Heterogeneous Graph Visualization')
plt.axis('off')
plt.show()
```
